```python
import librosa as lib
import numpy as np
from IPython.display import Audio
import matplotlib.pyplot as plt
```

```python
ruta = 'audio/'

# Asignamos, la frecuencia de sampleo
fs = 48000
# Duracion audio en seg
seg = 3

sig1, fs1 = lib.load(ruta+'aula-magna/voice-near.wav', sr=fs, duration=seg) # Se√±al Original
sig2, fs2 = lib.load(ruta+'aula-magna/pos1-ir.wav', sr=fs, duration=seg, offset=1) # IR aula magna
sig3, fs3 = lib.load(ruta+'reverb/reverb_1.wav', sr=fs, duration=seg, offset=6) # IR Reverberante
sig4, fs4 = lib.load(ruta+'anecoica/anecoica_2.wav', sr=fs, duration=seg, offset=3) # IR Anecoica
```

```python
Audio(data=sig2, rate=fs)
```

```python
print(fs1) # Chequeamos que este correcta la frecuencia de sampleo
```

```python
print(sig1.size/fs, sig2.size/fs, sig3.size/fs, sig4.size/fs) # Chequeamos el tama√±o de los audios
```

```python
# Convoluci√≥n Aula Magna
conv = np.convolve(sig1,sig2)
```

```python
Audio(data=conv, rate=fs)
```

```python
# Convolucion Reverberante
conv2 = np.convolve(sig1,sig3)
```

```python
Audio(data=conv2, rate=fs)
```

```python
# Convolucion Aneocoica
conv3 = np.convolve(sig1,sig4)
```

```python
Audio(data=conv3, rate=fs)
```

```python
fig1, ax1 = plt.subplots(2,2, figsize=(10, 6), layout='constrained');
fig1.suptitle('Convoluciones');
ax1[0,0].specgram(sig1, Fs=fs, cmap="magma");
ax1[0,1].specgram(sig2, Fs=fs, cmap="magma");
ax1[1,0].specgram(sig3, Fs=fs, cmap="magma");
ax1[1,1].specgram(sig4, Fs=fs, cmap="magma");
```
